# 다층 퍼셉트론 


다층 퍼셉트론(MLP, Multi-Layer Perceptron)은 신경망의 한 종류로서, 기본적인 인공신경망 구조 중 하나입니다. 이 모델은 여러 개의 레이어, 즉 층으로 구성되어 있으며, 각 레이어는 여러 개의 뉴런으로 이루어져 있습니다. MLP는 입력층, 하나 이상의 은닉층(hidden layers), 그리고 출력층으로 구성됩니다.

입력층(Input Layer): 이 층은 외부로부터 데이터를 받아들이는 역할을 합니다. 각 입력 뉴런은 데이터의 각 특성(feature)을 대표합니다.-

은닉층(Hidden Layers): 입력층과 출력층 사이에 위치하며, 데이터의 복잡한 패턴이나 특징을 학습하는 역할을 합니다. 은닉층의 뉴런 수와 층 수는 모델의 복잡성과 처리할 문제에 따라 달라질 수 있습니- 다.

출력층(Output Layer): 모델이 최종적으로 결정해야 할 출력을 제공합니다. 예를 들어, 분류 문제에서는 각 클래스에 대한 확률을 출력할 수 있- 습니다.

MLP의 각 뉴런은 활성화 함수를 통해 다음 층의 뉴런으로 신호를 전달합니다. 대표적인 활성화 함수로는 시그모이드(Sigmoid), 하이퍼볼릭 탄젠트(Tanh), 렐루(ReLU) 등이 있습니다.

학습 과정에서는 **역전파 알고리즘(Backpropagation)**이 사용됩니다. 이는 출력층에서 발생한 오차를 바탕으로 각 뉴런의 가중치를 조정하는 방법입니다. 학습의 목적은 이 오차를 최소화하는 것이며, 이를 위해 경사 하강법(Gradient Descent)과 같은 최적화 기법이 사용됩니다.

MLP는 간단한 XOR 문제 해결부터 복잡한 이미지 분류, 음성 인식 등 다양한 문제에 활용될 수 있습니다. 

소프트맥스 함수(Softmax function)는 주로 다중 클래스 분류 문제에서 사용되는 활성화 함수로서, 모델의 출력을 확률 분포로 변환하는 데 사용됩니다. 이 함수는 각 클래스에 대한 출력 값을 0과 1 사이의 확률 값으로 변환하며, 모든 출력 값의 합이 1이 되도록 보장합니다. 따라서 소프트맥스 함수의 출력은 각 클래스에 속할 확률을 나타내며, 가장 높은 확률을 가진 클래스가 예측 결과로 선택됩니다.

소프트맥스 함수는 주로 신경망의 마지막 층에서 사용되어, 각 클래스의 예측 확률을 계산하는 데 쓰이며, 이는 분류 작업의 결과를 직관적으로 해석할 수 있게 합니다.

```python
from sklearn.datasets import load_iris
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report


# 데이터 로드
iris = load_iris()
X = iris.data  # 특성 데이터
y = iris.target  # 타겟 레이블

# 데이터를 학습 세트와 테스트 세트로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)


# MLP 분류기 인스턴스 생성
mlp = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, alpha=1e-4,
                    solver='sgd', verbose=10, random_state=1,
                    learning_rate_init=.1)

# 모델 학습
mlp.fit(X_train, y_train)


# 테스트 세트에서 예측
y_pred = mlp.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print("정확도:", accuracy)

# 세부 성능 평가
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

```python
from sklearn.datasets import load_digits
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report


# 데이터 로드
digits = load_digits()
X = digits.data  # 특성 데이터
y = digits.target  # 타겟 레이블

# 데이터를 학습 세트와 테스트 세트로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)


# MLP 분류기 인스턴스 생성
mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=1000, alpha=1e-4,
                    solver='sgd', verbose=10, random_state=1,
                    learning_rate_init=.01)

# 모델 학습
mlp.fit(X_train, y_train)

# 테스트 세트에서 예측
y_pred = mlp.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print("정확도:", accuracy)

# 세부 성능 평가
print(classification_report(y_test, y_pred))

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline




# 데이터 로드
california = fetch_california_housing()
X = california.data
y = california.target

# 데이터를 학습 세트와 테스트 세트로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# 모델 리스트 생성
models = {
    "Decision Tree": DecisionTreeRegressor(random_state=1),
    "MLP Regressor": make_pipeline(StandardScaler(), MLPRegressor(hidden_layer_sizes=(50,), max_iter=1000, random_state=1)),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=1),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=1)
}

# 각 모델에 대한 교차 검증 점수 계산
scores = {name: cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error') for name, model in models.items()}


# MSE 점수를 양수로 변환하고, 평균과 표준편차 계산
mean_scores = {name: -np.mean(score) for name, score in scores.items()}
std_scores = {name: np.std(score) for name, score in scores.items()}

# 성능 시각화
fig, ax = plt.subplots()
names = list(mean_scores.keys())
values = list(mean_scores.values())
errors = list(std_scores.values())
ax.barh(names, values, xerr=errors, align='center', color='blue', ecolor='red')
ax.set_xlabel('Mean Squared Error')
ax.set_title('Model Comparison on California Housing Dataset')
plt.show()

```
