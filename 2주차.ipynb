# 데이터 과학 프로젝트 가이드

이 가이드는 데이터 과학 프로젝트를 진행할 때 따를 수 있는 단계별 절차를 설명합니다. 주요 단계로는 데이터셋 준비하기, 데이터셋 분할하기, 데이터 전처리 및 모델 학습, 예측 데이터 생성과 평가, 그리고 하이퍼 파라미터 튜닝 및 반복이 있습니다.

## 1. 데이터셋 준비하기

데이터셋 준비는 프로젝트의 출발점입니다. 이 단계에서는 필요한 데이터를 수집하고, 데이터의 초기 형태를 분석합니다. 데이터 수집 방법은 다양하며, 공개 데이터셋을 사용하거나, API를 통해 데이터를 수집하거나, 웹 스크래핑을 통해 정보를 추출할 수 있습니다.

## 2. 데이터셋 분할하기

데이터셋을 학습, 검증, 테스트 데이터셋으로 분할합니다. 이 과정은 모델이 새로운 데이터에 대해 얼마나 잘 일반화할 수 있는지 평가하는 데 필수적입니다.

- **학습 데이터셋**: 모델 학습에 사용됩니다.
- **검증 데이터셋**: 학습 중 모델 성능을 평가하고 하이퍼파라미터를 조정하는 데 사용됩니다.
- **테스트 데이터셋**: 모델 학습을 마친 후, 모델의 최종 성능을 평가하는 데 사용됩니다.

## 3. 데이터 전처리 및 모델 학습

데이터 전처리 단계에서는 누락된 값 처리, 범주형 데이터의 수치화, 특성 스케일링 등을 포함한 다양한 기술을 적용하여 데이터를 모델이 처리할 수 있는 형태로 변환합니다. 이후 적절한 머신러닝 알고리즘을 선택하여 학습 데이터셋을 사용해 모델을 학습시킵니다.

## 4. 예측 데이터 생성과 평가

학습된 모델을 사용하여 검증 데이터셋과 테스트 데이터셋에 대한 예측을 수행합니다. 이 단계에서는 모델의 성능을 평가하기 위해 정확도, 정밀도, 재현율, F1 점수와 같은 다양한 평가 지표를 사용할 수 있습니다.

## 5. 하이퍼 파라미터 튜닝 및 반복

모델의 성능을 최적화하기 위해 하이퍼 파라미터 튜닝을 수행합니다. 그리드 탐색, 랜덤 탐색, 베이지안 최적화와 같은 방법을 사용할 수 있습니다. 최적의 하이퍼 파라미터를 찾은 후, 변경된 설정으로 모델을 재학습시키고 성능을 다시 평가합니다. 이 과정은 모델의 성능이 만족스러울 때까지 반복될 수 있습니다.

이러한 단계를 반복하며 모델의 성능을 지속적으로 개선해 나가는 것이 데이터 과학 프로젝트의 핵심입니다. 각 단계는 프로젝트의 목표와 사용할 데이터의 특성에 따라 조정될 수 있습니다.

```python
#정규 방정식으로 풀이 구현하기

import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

X, y = load_diabetes(return_X_y=True, as_frame=False)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=1234)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

n_train, n_test = X_train.shape[0], X_test.shape[0]
```

```python
X_train = np.append(np.ones((n_train, 1)), X_train, axis=1)
```

```python
w = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
```

```python
y_pred_train = X_train @ w
print(f'학습 데이터셋 MAE:{np.abs(y_pred_train - y_train).mean(): .3f}')

X_test = np.append(np.ones((n_test, 1)), X_test, axis=1)
y_pred = X_test @ w
print(f'테스트 데이터셋 MAE:{np.abs(y_pred - y_test).mean(): .3f}')
```

```python
import numpy as np

np.random.seed(1234)
X = np.random.randn(1000, 10)
Y = np.linspace(0, 1, len(X))
```

```python
print(f'X의 조건수: {np.linalg.cond(X)}')
print(f'X.T @ X의 조건수: {np.linalg.cond(X.T @ X)}')
print(f'X 조건수 제곱과 X.T @ X 조건수의 비율: {np.linalg.cond(X)**2 / np.linalg.cond(X.T @ X)}')
```

```python
w_normal = (np.linalg.inv(X.T @ X)) @ X.T @ Y
w_svd, _, _, _ = np.linalg.lstsq(X, Y, rcond=None)

print(f'정규 방정식 풀이의 파라미터:\n{w_normal}\n')
print(f'SVD-OLS 풀이의 파라미터:\n{w_svd}\n')
print(f'두 해가 허용 오차 범위에서 같은지를 판별한 결과: {np.allclose(w_normal, w_svd)}')
```

```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

X, y = load_diabetes(return_X_y=True, as_frame=False)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=1234)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

n_train, n_test = X_train.shape[0], X_test.shape[0]

X_train = np.append(np.ones((n_train, 1)), X_train, axis=1)
X_test = np.append(np.ones((n_test, 1)), X_test, axis=1)
```

```python
u, s, vh = np.linalg.svd(X_train, full_matrices=False)

w = vh.T @ np.linalg.inv(np.diag(s)) @ u.T @ y_train
y_pred_train = X_train @ w
print(f'학습 데이터셋 MAE:{np.abs(y_pred_train - y_train).mean(): .3f}')

y_pred = X_test @ w
print(f'테스트 데이터셋 MAE:{np.abs(y_pred - y_test).mean(): .3f}')
```

```python
import numpy as np
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
%matplotlib inline

X, y = load_diabetes(return_X_y=True, as_frame=False)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=1234)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

n_train, n_test = X_train.shape[0], X_test.shape[0]

X_train = np.append(np.ones((n_train, 1)), X_train, axis=1)
X_test = np.append(np.ones((n_test, 1)), X_test, axis=1)

u, s, vh = np.linalg.svd(X_train, full_matrices=False)
print('Sigma 값:', s)
```

```python
num_trunc = 2

vh = vh[:-num_trunc, :]
s = s[:-num_trunc]
u = u[:, :-num_trunc]

w = vh.T @ np.linalg.inv(np.diag(s)) @ u.T @ y_train
y_pred_train = X_train @ w
print(f'학습 데이터셋 MAE:{np.abs(y_pred_train - y_train).mean(): .3f}')

y_pred = X_test @ w
print(f'테스트 데이터셋 MAE:{np.abs(y_pred - y_test).mean(): .3f}')
```

```python
train_maes = []
test_maes = []

num_truncs = range(X_train.shape[1])

for num_trunc in num_truncs:
    u, s, vh = np.linalg.svd(X_train, full_matrices=False)
    
    if num_trunc != 0:
        vh = vh[:-num_trunc, :]
        s = s[:-num_trunc]
        u = u[:, :-num_trunc]

    w = vh.T @ np.linalg.inv(np.diag(s)) @ u.T @ y_train
    y_pred_train = X_train @ w
    
    train_maes.append(np.abs(y_pred_train - y_train).mean())
    
    y_pred = X_test @ w
    
    test_maes.append(np.abs(y_pred - y_test).mean())
    
plt.plot(num_truncs, train_maes, label='Train Set MAE')
plt.plot(num_truncs, test_maes, label='TEST Set MAE')
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression

X, y = load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=1234)
```

```python
reg = LinearRegression()
reg = reg.fit(X_train, y_train)
y_pred_train = reg.predict(X_train)

print(f'학습 데이터셋 MAE:{np.abs(y_pred_train - y_train).mean(): .3f}')

y_pred = reg.predict(X_test)
print(f'테스트 데이터셋 MAE:{np.abs(y_pred - y_test).mean(): .3f}')
```
