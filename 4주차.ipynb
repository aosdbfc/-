# 분석 모형 구축

### 1. 데이터 모델 구축 기법

- 빅데이터 분석에서는 분석하고자 하는 데이터를 활용하여 목적과 형태에 맞는 분석 모형을 선정한다.
- 데이터 마이닝 및 머신러닝을 이용한 데이터 모델링은 분석 데이터를 활용하여 분류, 예측, 군집 등의 분석모형을 선정하여 개발할 수 있다. 
- 빅데이터 분석모형 선정을 위해서는 먼저 데이터 모델 구축 기법들에 대해서 살펴볼 필요가 있다.

### 가. 통계 분석 

- 통계학에서 전형적인 데이터 모델 구축 기법에서는 회귀 분석, 로지스틱 회귀분석, 판별분석, 주성분 분석 등이 있다.

- <회귀분석> 
    - 종속변수에 대한 독립변수의 선형함수 관계로부터 새로운 값에 대해서 종속변수의 값을 예측
    - 인과성을 밝히는 데도 사용할 수 있다.
    
- <로지스틱 회귀분석>
    - 설명변수값이 주어졌을 때 목표변수 값이 특정 부류에 속할 확률이 로지스틱 함수 형태를 따르는 것을 이용하여 개별 관측치들이 어느 집단에 분류되는지 예측한다. 
    
- <판별분석>
    - 종속변인이 둘 혹은 그 이상의 집단으로 구성되어 있을 때 여러 개의 독립변인으로 집단 관측치를 판별 혹은 예측한다.
- <주성분 분석>
    - 데이터에 여러 변수들이 있을 때, 서로 상관관계가 높은 변수들의 선형결합으로 만들어진 주성분이라는 새로운 변수를 만들어 변수들을 요약하고 축소한다.

### 나. 데이터 마이닝

- 분류
- 예측 
- 연관 : 일련의 거래 또는 사건들 간의 규칙을 발견하여 조건과 반응의 형태(if-then구조)로 분석 결과의 연관성을 파악 
- 군집화 : 주어진 데이터들의 특성을 고려해 비슷한 특성을 가진 데이터들을 소그룹으로 묶는 방법 : 보통 비지도 학습 


### <기법과 알고리즘>

- 분류 : 로지스틱 회귀, 의사결정 나무, 서포트 벡터 머신, 나이브 베이즈, K-최근접 이웃, 인공신경망
- 예측 : 회귀분석, 장바구니분석, 시계열 분석, 의사결정나무, 인공신경망, K-최근접 이웃
- 연관 : 연관성 분석, 순차패턴 분석
- 군집화 : 군집 분석, K-means 클러스터링, 기대최대화

#### <지도학습(Supervised Learning)>

지도학습은 입력 데이터(X)와 그에 대응하는 출력 데이터(Y)의 쌍으로 구성된 데이터 세트를 사용합니다. 모델은 이 데이터를 학습하여, 새로운 입력 데이터에 대한 적절한 출력을 예측하려고 합니다. 지도학습의 주요 목표는 입력과 출력 사이의 관계를 모델링하는 것이며, 이를 통해 분류(Classification) 또는 회귀(Regression) 문제를 해결할 수 있습니다.

- 분류(Classification): 출력 변수가 카테고리형 데이터(예: 이메일이 스팸인지 아닌지)일 경우 사용합니다.
- 회귀(Regression): 출력 변수가 연속적인 수치(예: 집의 가격)일 경우 사용합니다.

#### <비지도학습(Unsupervised Learning)>

비지도학습은 출력 데이터(Y) 없이 입력 데이터(X)만을 사용하여 데이터의 구조나 패턴을 찾는 학습 방법입니다. 비지도학습의 주요 목표는 데이터 내의 숨겨진 패턴이나 구조를 발견하는 것이며, 주로 군집화나 연관 규칙 발견과 같은 작업에 사용됩니다.

- 군집(Clustering): 데이터를 유사성에 따라 여러 그룹으로 나누는 과정입니다. 군집화는 시장 세분화, 사회 네트워크 분석, 검색 엔진 최적화 등 다양한 분야에서 사용됩니다.
- 연관 규칙(Association Rules): 데이터 항목 간의 연관성을 찾는 규칙을 발견하는 것입니다. 연관 규칙은 소비자 구매 패턴 분석, 상품 추천 시스템, 재고 관리 등에 사용됩니다. ex) 어피리오리, 빈출패턴 성장 

### 데이터 분할

- 데이터 분할은 훈련용, 평가용, 검증용 데이터로 분류합니다. 
- 필요와 데이터의 양에 따라서 데이터를 분할한다는 점을 기억하고 각 특징을 파악해야합니다.
- 데이터 분할 시 유의할 점은 분할 전에 데이터의 0,1(양성 혹은 음성) 클래스가 훈련 데이터셋이나 평가 데이터셋의 어느 한쪽에 몰리지 않도록 골고루 섞어야합니다.

< 데이터의 양이 충분하지 않거나 입력 변수에 대한 설명이 충분한 경우>
- 홀드아웃 방법
    - 주어진 데이터를 랜덤하게 두 개의 데이터로 구분하여 사용하는 방법으로 주로 훈련데이터와 평가데이터를 분리하여 사용한다
- 교차 검증 방법 
    - K-겹 교차 검증에서는 데이터를 K개의 동일한 크기를 가진 부분 집합으로 나눕니다. 그 후, K-1개의 부분 집합을 합쳐서 모델을 훈련하고, 남은 1개의 부분 집합으로 모델을 검증합니다. 이 과정을 K번 반복하여, 각 부분 집합이 정확히 한 번씩 검증용 데이터로 사용됩니다. 마지막에는 K번의 검증 결과를 평균내어 모델의 성능을 평가합니다.

## 6장 의사결정 알고리즘 - decision tree

```python
# 결정 트리 분류 모델 구현하기

import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

df = load_iris(as_frame=True)['frame']
df = df[df['target'] <= 1]
df = df.sample(frac=1, random_state=1234)

train, test = df.iloc[:75, :], df.iloc[75:, :]
X_train, y_train = train.iloc[:, :-1], train.iloc[:, -1]
X_test, y_test = test.iloc[:, :-1], test.iloc[:, -1]
X_cols, y_col = train.columns.tolist()[:-1], train.columns.tolist()[-1]
```

```python
max_depth = 4
min_samples_split = 4
```

```python
def eval_gini(left, right, y_col):
    gini, n1, n2 = 0, len(left), len(right)
    
    score = 1 - ((left[y_col].value_counts())**2).sum() / (n1**2)
    gini += score * n1 / (n1 + n2)

    score = 1 - ((right[y_col].value_counts())**2).sum() / (n2**2)
    gini += score * n2 / (n1 + n2)

    return gini
```

```python
def eval_class(df, y_col):
    return df[y_col].value_counts().sort_values(ascending=False).index[0]
```

```python
tree = []
for i in range(0, 2**(max_depth + 1)):
    tree.append(dict({'struct': None}))

tree[1]['struct'] = train
```

```python
for i in range(1, len(tree)):

    # 트리가 끝난 부분은 통과
    if not isinstance(tree[i]['struct'], pd.DataFrame): continue
    
    # 최대 깊이 도달하였으면 클래스를 출력
    if i >= 2**max_depth:
        tree[i]['struct'] = eval_class(tree[i]['struct'], y_col)
        continue
    
    data = tree[i]['struct']
    a, b, c, d, e = '', float('inf'), float('inf'), None, None
    
    # 최고의 피처와 임계값을 찾은 후 그 기준으로 분할
    for X_col in X_cols: 
        vals = np.sort(data[X_col].unique())
        for val in vals[1:]:
            left, right = data[data[X_col] < val], data[data[X_col] >= val]

            gini = eval_gini(left, right, y_col)

            if gini < c:
                a, b, c, d, e = X_col, val, gini, left, right
    
    tree[i]['col'] = a
    tree[i]['val'] = b
    if len(d) >= min_samples_split:
        tree[i << 1]['struct'] = d
    else:
        tree[i << 1]['struct'] = eval_class(e, y_col)
    if len(e) >= min_samples_split:
        tree[(i << 1) + 1]['struct'] = e
    else:
        tree[(i << 1) + 1]['struct'] = eval_class(e, y_col)
```

```python
cnt = 0
for i in range(len(X_train)):
    row = X_train.iloc[i]
    ind = 1
    node = tree[ind]
    while isinstance(node['struct'], pd.DataFrame):
        if row[node['col']] < node['val']:
            ind = ind << 1
        else:
            ind = (ind << 1) + 1
        node = tree[ind]

    y_pred = node['struct']
    cnt += (y_pred == y_train.iloc[i]) * 1
    
print(f'학습 데이터셋 정확도:{cnt / len(y_train) * 100: .2f}%')
```

```python
cnt = 0
for i in range(len(X_test)):
    row = X_test.iloc[i]
    ind = 1
    node = tree[ind]
    while isinstance(node['struct'], pd.DataFrame):
        if row[node['col']] < node['val']:
            ind = ind << 1
        else:
            ind = (ind << 1) + 1
        node = tree[ind]

    y_pred = node['struct']
    cnt += (y_pred == y_test.iloc[i]) * 1
    
print(f'테스트 데이터셋 정확도:{cnt / len(y_test) * 100: .2f}%')
```

# 그리드 서치로 결정 트리의 하이퍼파라미터 튜닝하기

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X, y = load_breast_cancer(return_X_y=True, as_frame=False)

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=1234)
X_train, X_val, y_train, y_val = train_test_split(X_train,
                                                  y_train,
                                                  test_size=0.33,
                                                  random_state=1234)
```

```python
max_depths = [None, 3, 6]
min_samples_splits = [2, 3, 4]
max_featuress = ['sqrt', 'log2', None]
```

```python
from itertools import product

best_max_depth = None
best_min_samples_split = None
best_max_features = None

best_acc = 0

for max_depth, min_samples_split, max_features in product(max_depths, min_samples_splits, max_featuress):
    clf = DecisionTreeClassifier(random_state=1234,
                                 max_depth=max_depth,
                                 min_samples_split=min_samples_split,
                                 max_features=max_features)
    y_pred = clf.fit(X_train, y_train).predict(X_val)
    
    acc = (y_pred == y_val).mean()
    if acc > best_acc:
        best_acc = acc
        best_max_depth = max_depth
        best_min_samples_split = min_samples_split
        best_max_features = max_features

print('max_depth:', best_max_depth, '\nmin_samples_split:', 
      best_min_samples_split, '\nmax_features:', best_max_features)
```

```python
clf = DecisionTreeClassifier(random_state=1234,
                             max_depth=best_max_depth,
                             min_samples_split=best_min_samples_split,
                             max_features=best_max_features)

y_pred = clf.fit(X_train,y_train).predict(X_test)
print(f'정확도:{(y_pred == y_test).mean() * 100: .2f}%')
```

# 7장 - 랜덤 포레스트 모델

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

df = load_iris(as_frame=True)['frame']
df = df[df['target'] <= 1]
df = df.sample(frac=1, random_state=1234)

train, test = df.iloc[:75, :], df.iloc[75:, :]
X_train, y_train = train.iloc[:, :-1], train.iloc[:, -1]
X_test, y_test = test.iloc[:, :-1], test.iloc[:, -1]
X_cols, y_col = train.columns.tolist()[:-1], train.columns.tolist()[-1]

def eval_gini(left, right, y_col):
    gini, n1, n2 = 0, len(left), len(right)
    score = 1 - ((left[y_col].value_counts())**2).sum() / (n1**2)
    gini += score * n1 / (n1 + n2)
    score = 1 - ((right[y_col].value_counts())**2).sum() / (n2**2)
    gini += score * n2 / (n1 + n2)
    return gini

def eval_class(df, y_col):
    return df[y_col].value_counts().sort_values(ascending=False).index[0]
```

```python
def base_tree(train, X_cols, y_col, max_depth=4, min_samples_split=4):
    tree = []
    for i in range(0, 2**(max_depth + 1)):
        tree.append(dict({'struct': None}))

    tree[1]['struct'] = train

    for i in range(1, len(tree)):
        if not isinstance(tree[i]['struct'], pd.DataFrame): continue
        if i >= 2**max_depth:
            tree[i]['struct'] = eval_class(tree[i]['struct'], y_col)
            continue
        data = tree[i]['struct']
        a, b, c, d, e = '', float('inf'), float('inf'), None, None
        for X_col in X_cols: 
            vals = np.sort(data[X_col].unique())
            for val in vals[1:]:
                left, right = data[data[X_col] < val], data[data[X_col] >= val]
                gini = eval_gini(left, right, y_col)
                if gini < c: a, b, c, d, e = X_col, val, gini, left, right

        tree[i]['col'] = a
        tree[i]['val'] = b
        if len(d) >= min_samples_split: tree[i << 1]['struct'] = d
        else: tree[i << 1]['struct'] = eval_class(e, y_col)
            
        if len(e) >= min_samples_split: tree[(i << 1) + 1]['struct'] = e
        else: tree[(i << 1) + 1]['struct'] = eval_class(e, y_col)
    return tree
```

```python
max_samples = 0.3
forest = []
for i in range(9):
    train0 = train.sample(frac=max_samples, random_state=1234)
    forest.append(base_tree(train0, X_cols, y_col))
```

```python
cnt = 0
for i in range(len(X_test)):
    row = X_test.iloc[i]
    y_pred = [0, 0]
    for tree in forest:
        ind = 1
        node = tree[ind]
        while isinstance(node['struct'], pd.DataFrame):
            if row[node['col']] < node['val']: ind = ind << 1
            else: ind = (ind << 1) + 1

            node = tree[ind]

        y_pred_tree = node['struct']
        y_pred[y_pred_tree] += 1
    
    if y_pred[0] > y_pred[1]: y_pred = 0
    else: y_pred = 1
    
    cnt += (y_pred == y_test.iloc[i]) * 1
    
print(f'테스트 데이터셋 정확도:{cnt / len(y_test) * 100: .2f}%')
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline

X, y = make_classification(n_samples=500,
                           n_features=25,
                           n_clusters_per_class=1,
                           n_informative=15,
                           random_state=123)
```

```python
ensemble_clfs = [(0, "max_features = 'sqrt'",
                  RandomForestClassifier(warm_start=True,
                                         oob_score=True,
                                         max_features="sqrt",
                                         random_state=123)),
                 (1, "max_features = 'log2'",
                  RandomForestClassifier(warm_start=True,
                                         max_features="log2",
                                         oob_score=True,
                                         random_state=123)),
                 (2, "max_features = None",
                  RandomForestClassifier(warm_start=True,
                                         max_features=None,
                                         oob_score=True,
                                         random_state=123))] 
```

```python
min_estimators = 15
max_estimators = 150

oob_scores = [[], [], []]

for ind, label, clf in ensemble_clfs:
    for i in range(min_estimators, max_estimators + 1, 5):
        clf.set_params(n_estimators=i)
        clf.fit(X, y)
        oob_scores[ind].append((i, clf.oob_score_))

for i in range(len(ensemble_clfs)):
    xs, ys = zip(*oob_scores[i])
    plt.plot(xs, ys, label=ensemble_clfs[i][1])

plt.xlim(min_estimators, max_estimators)
plt.xlabel("n_estimators")
plt.ylabel("OOB Score")
plt.legend(loc="best")
```

# 랜덤 포레스트 분류 모델의 피처 중요도

```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline

X, y = make_classification(n_samples=1000,
                           n_features=10,
                           n_informative=3,
                           n_redundant=0,
                           n_classes=2,
                           random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    stratify=y,
                                                    random_state=42)
```

```python
feature_names = [f"feature {i}" for i in range(X.shape[1])]
forest = RandomForestClassifier(random_state=0).fit(X_train, y_train)
```

```python
start_time = time.time()
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
print(f'MDI 기반 피처 중요도 산출 소요 시간:{time.time() - start_time: .3f}초')
```

```python
forest_importances = pd.Series(importances, index=feature_names)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

```python
from sklearn.inspection import permutation_importance # 나열 기반 
start_time = time.time()
result = permutation_importance(forest,
                                X_test,
                                y_test,
                                n_repeats=10,
                                random_state=0)
print(f'퍼뮤테이션 기반 피처 중요도 산출 소요 시간:{time.time() - start_time: .3f}초')

forest_importances = pd.Series(result.importances_mean, index=feature_names)
```

```python
fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=result.importances_std, ax=ax)
ax.set_title('Feature importances using permutation on full model')
ax.set_ylabel('Mean accuracy decrease')
fig.tight_layout()
plt.show()
```

# 8장 부스팅 계열 모델 - Boosting

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import numpy as np

df = load_breast_cancer(as_frame=True)['frame'].sample(
    frac=1, random_state=12).reset_index(drop=True)

def transfer_y(x):
    if x == 0: return -1
    return x

df['target'] = df['target'].apply(lambda x: transfer_y(x))

train = df.iloc[:400, :]
test = df.iloc[400:, :]

X_train, y_train = train.drop(['target'], axis=1), train['target']
X_test, y_test = test.drop(['target'], axis=1), test['target']
```

```python
clfs = []
alphas = []

pmf = np.ones(len(train)) / len(train)

for i in range(10):
    sampled = train.sample(frac=1,
                           replace=True,
                           weights=pmf,
                           random_state=1234).reset_index(drop=True)
    X_sampled = sampled.drop(['target'], axis=1)
    y_sampled = sampled['target']
    
    clf = DecisionTreeClassifier(max_depth=1,
                                 random_state=1234).fit(X_sampled, y_sampled)
    y_pred = clf.predict(X_sampled)
    eps = (y_pred != y_sampled).mean()
    if eps == 0:
        print(f'오분류율이 0%에 도달하여 학습 조기 종료: 트리의 수 = {len(clfs)}')
        break
    if eps >= 0.5:
        print('오분류율 50% 이상이므로 학습 종료')
        break
    alpha = 1 / 2 * np.log((1 - eps) / eps) 

    pmf = pmf * np.exp(-alpha * y_sampled * y_pred)
    pmf = pmf / pmf.sum()
    
    clfs.append(clf)
    alphas.append(alpha)
    if len(clfs) == 10:
        print(f'i = {i}에서 이터레이션 정상 종료')
        break
```

```python
for clf in clfs:
    print(
        f'분기에 사용된 피처: {clf.tree_.feature[0]}, 분기 기준값:{clf.tree_.threshold[0]: .4f}'
    )
```

```python
y_train_pred = np.zeros(len(y_train))

for alpha, clf in zip(alphas, clfs):
    y_train_pred += alpha * clf.predict(X_train)

y_train_pred = (y_train_pred > 0) * 1
y_train_pred[y_train_pred == 0] = -1

print(f'트리의 수 = {len(clfs)}인 에이다부스트 모델의 학습 데이터셋 기준 정확도: {(y_train_pred == y_train).mean() * 100: .2f}%')

y_pred = np.zeros(len(y_test))

for alpha, clf in zip(alphas, clfs):
    y_pred += alpha * clf.predict(X_test)

y_pred = (y_pred > 0) * 1
y_pred[y_pred == 0] = -1

print(f'트리의 수 = {len(clfs)}인 에이다부스트 모델의 테스트 데이터셋 기준 정확도:{(y_pred == y_test).mean() * 100: .2f}%')
```

# 결정 트리와 에이다부스트 비교

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

rng = np.random.RandomState(1)
X = np.linspace(0, 6, 100)[:, np.newaxis]
y = np.sin(X).ravel() + np.sin(6 * X).ravel() + rng.normal(0, 0.1, X.shape[0])

colors = sns.color_palette("colorblind")
plt.scatter(X, y, color=colors[0])
plt.xlabel("data"), plt.ylabel("target"), plt.title("Sample dataset")
```

```python
from sklearn.tree import DecisionTreeRegressor

regr_1 = DecisionTreeRegressor(max_depth=4).fit(X, y)
y_1 = regr_1.predict(X)

plt.scatter(X, y, color=colors[0], label="training samples")
plt.plot(X, y_1, color=colors[1], label="DT Regressor", linewidth=2)
plt.xlabel("data"), plt.ylabel("target"), plt.title("DT Regressor"),
plt.legend()
```

```python
from sklearn.ensemble import AdaBoostRegressor

regr_2 = AdaBoostRegressor(DecisionTreeRegressor(max_depth=4),
                           n_estimators=300,
                           random_state=rng).fit(X, y)
y_2 = regr_2.predict(X)

plt.scatter(X, y, color=colors[0], label="training samples")
plt.plot(X, y_2, color=colors[2], label="Adaboost", linewidth=2)
plt.xlabel("data"), plt.ylabel("target"), plt.title("Adaboost"), plt.legend()
```

# 부스팅 트리 회귀 모델 구현하기

```python
from sklearn.datasets import load_diabetes
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import numpy as np

df = load_diabetes(as_frame=True)['frame']
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=1234)
```

```python
y_train2 = y_train.copy()

trees = []
for i in range(100):
    reg = DecisionTreeRegressor(max_depth=4, 
                                random_state=1234).fit(X_train, y_train2)
    y_pred = reg.predict(X_train)
    y_train2 = y_train2 - y_pred
    trees.append(reg)
```

```python
y_pred = np.zeros(len(y_test))
for tree in trees:
    y_pred += tree.predict(X_test)

print(f'테스트 데이터셋 기준 MAE:{(np.abs(y_pred-y_test)).mean(): .2f}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor

reg = GradientBoostingRegressor(n_estimators=100,
                                max_depth=4,
                                random_state=1234)
y_pred = reg.fit(X_train, y_train).predict(X_test)

print(f'GradientBoostingRegressor 테스트 데이터셋 기준 MAE:{(np.abs(y_pred - y_test)).mean(): .2f}')

y_pred = DecisionTreeRegressor(random_state=1234).fit(X_train,
                                                      y_train).predict(X_test)
print(f'DecisionTreeRegressor 테스트 데이터셋 기준 MAE:{(np.abs(y_pred - y_test)).mean(): .2f}')
```

```python
# lime_text 모듈을 통한 텍스트 모델의 사후 분석

import pandas as pd
import numpy as np
import urllib.request
from konlpy.tag import Okt

urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt",
    filename="ratings_train.txt")
urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt",
    filename="ratings_test.txt")
```

```python
train = pd.read_table('ratings_train.txt').sample(
    frac=0.1, random_state=1234).reset_index()
train = train.dropna(how='any')

test = pd.read_table('ratings_test.txt').sample(
    frac=0.1, random_state=1234).reset_index()
test = test.dropna(how='any')

y_train = train['label']
y_test = test['label']

print('학습 데이터셋 크기:', train.shape)
print('테스트 데이터셋 크기:', test.shape)
print('y_train == 1의 비율:', y_train.mean())

train.head()
```

```python
okt = Okt()

def get_nouns(x):
    return ' '.join(okt.nouns(x))

train['document_nouns'] = train['document'].apply(lambda x: get_nouns(x))
test['document_nouns'] = test['document'].apply(lambda x: get_nouns(x))

train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

text_train = train['document_nouns']
text_test = test['document_nouns']

tfidf_vectorizer = TfidfVectorizer(min_df = 0.001, max_df = 0.3)
tfidf_vectorizer.fit(text_train)

X_train_tfidf = tfidf_vectorizer.transform(text_train)
X_test_tfidf = tfidf_vectorizer.transform(text_test)

print(X_train_tfidf.shape, X_test_tfidf.shape)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(random_state=1234)

clf.fit(X_train_tfidf, y_train)

print("Train Accuracy:", clf.score(X_train_tfidf, y_train))
print("Test Accuracy:", clf.score(X_test_tfidf, y_test))
```
